## Creamos el dataframe para las precipitaciones

### Creamos el diccionario de estados_cielo para sustituir los estados del cielo en la tabla de precipitaciones

```python
df_estados_cielo = pd.DataFrame(df_aemet['cielo'].unique())
```

```python
dict_reemplazo_estados_cielo = dict(df_estados_cielo.T.iloc[0])
dict_reemplazo_estados_cielo = {v: k for k, v in dict_reemplazo_estados_cielo.items()}
```

```python
# with open('datos/jsons/dict_remplazo_estados_cielo.json', 'w') as archivo_json:
#     json.dump(dict_reemplazo_estados_cielo, archivo_json)
```

```python
df_cielo = pd.DataFrame()
df_cielo['estado'] = dict_reemplazo_estados_cielo.values()
df_cielo['id'] = dict_reemplazo_estados_cielo.keys()
```

```python
#df_cielo.to_csv('datos/tratados/df_estados_cielo.csv')
```

### Reemplazamos los valores del cielo por sus ids de la futura planta

```python
df_aemet['cielo'] = df_aemet['cielo'].replace(dict_reemplazo_estados_cielo)
```

### Cambiamos el valor de las fechas a tipo fecha

```python
df_aemet['fecha'] = pd.to_datetime(df_aemet['fecha'], format='%Y-%m-%d %H:%M:%S')
```
